```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from utils import get_train_data, BASE_DIR
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
```

### 13 classes

```python
X_train, y_train, X_val, y_val = get_train_data(skip_midpoints=True, all_labels=True)
```

```python
LABELS = [
    'left jab weak',
    'right jab weak',
    'left hook weak',
    'right hook weak',
    'left uper weak',
    'right uper weak',
    
    'left jab strong',
    'right jab strong',
    'left hook strong',
    'right hook strong',
    'left uper strong',
    'right uper strong'
]
```

```python
y_train = [i for i in y_train if i != 0]
y_val = [i for i in y_val if i != 0]
```

```python
def plot_hist(x, labels, hist_name, img_name):
    sns.set()
    plt.hist(x, bins=[i + 0.5 for i in range(len(LABELS) + 1)])
    plt.title(hist_name, fontsize=18)
    plt.xticks([i + 1 for i in range(len(LABELS))], labels=LABELS, rotation=45)
    plt.tight_layout()
    plt.savefig(BASE_DIR.joinpath('imgs', f'{img_name}.png'))
    plt.show()
```

```python
plot_hist(y_train, LABELS, 'Histogram of training labels', 'hist_of_labels_13_train')
```

```python
plot_hist(y_val, LABELS, 'Histogram of validation labels', 'hist_of_labels_13_val')
```
